# Image Extraction

```python
def extract_tar_gz(tar_path, path_to_extract):
    # avoid duplicates -> check if the directory already exists
    dir_name = os.path.basename(tar_path).replace('.tar.gz', '')
    output_dir = os.path.join(path_to_extract, dir_name)

    # only extract images if that directory doesn't exist
    if not os.path.exists(output_dir):
        with tarfile.open(tar_path) as tar:
            tar.extractall(path=path_to_extract)
            print(f'Extracted {tar_path}')
    else:
        print(f'Skipped {tar_path}, already extracted')

# paths to .tar.gz files
tar_files = ['data/images_01.tar.gz', 'data/images_02.tar.gz',
             'data/images_03.tar.gz', 'data/images_04.tar.gz',
             'data/images_05.tar.gz', 'data/images_06.tar.gz',
             'data/images_07.tar.gz', 'data/images_08.tar.gz',
             'data/images_09.tar.gz', 'data/images_10.tar.gz',
             'data/images_11.tar.gz', 'data/images_12.tar.gz',]

# directory to store extracted images
extracted_images_dir = 'extracted_images/images'

# create create if doesn't exist
os.makedirs(extracted_images_dir, exist_ok=True)

# extract files from .tar.gz archives
for tar_file in tar_files:
    extract_tar_gz(tar_file, extracted_images_dir)
```

# Data Load and Prep

```python
# 14 target disease categories
disease_categories = ['Atelectasis',
                      'Cardiomegaly',
                      'Effusion',
                      'Infiltration',
                      'Mass',
                      'Nodule',
                      'Pneumonia',
                      'Pneumothorax',
                      'Consolidation',
                      'Edema',
                      'Emphysema',
                      'Fibrosis',
                      'Pleural_Thickening',
                      'Hernia']

# transforms for image preprocessing
transform = transforms.Compose([
    transforms.Resize((128, 128)),  # resize the images
    transforms.RandomRotation(degrees=15),  # rotate randomly between -15 and 15 degrees
    transforms.RandomResizedCrop(size=128, scale=(0.9, 1.0)),  # random crop with resize
    transforms.RandomHorizontalFlip(),  # randomly flip horizontally
    transforms.ToTensor()])  # convert to PyTorch tensors

# make custom Dataset class for chest x-ray images
class ChestXRayDataset(Dataset):
    # constructor
    def __init__(self, dataframe, image_dir, transform=None):
        self.dataframe = dataframe
        self.image_dir = image_dir
        self.transform = transform

    # length method
    def __len__(self):
        return len(self.dataframe)

    # get item method
    def __getitem__(self, idx):
        img_name = os.path.join(self.image_dir, self.dataframe.iloc[idx, 0])
        image = Image.open(img_name).convert('RGB')
        
        # create multi-label target vector
        label = np.zeros(len(disease_categories), dtype=float)
        for disease in self.dataframe.iloc[idx, 1].split('|'):
            if disease in disease_categories:
                label[disease_categories.index(disease)] = 1
        
        # apply transforms
        if self.transform:
            image = self.transform(image)
        
        return image, torch.from_numpy(label).float()

# create dataset and dataloaders
dataset = ChestXRayDataset(metadata, extracted_images_dir, transform=transform)
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size
train_dataset, val_dataset = torch.utils.data.random_split(dataset, [train_size, val_size])
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
```

# Define CNN Model

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# simple CNN model
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.fc1 = nn.Linear(64 * 32 * 32, 64)
        self.fc2 = nn.Linear(64, len(disease_categories))
        self.pool = nn.MaxPool2d(2, 2)
        self.dropout = nn.Dropout(0.5)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        x = self.pool(self.relu(self.conv1(x)))
        x = self.pool(self.relu(self.conv2(x)))
        x = torch.flatten(x, 1)  # flatten tensors
        x = self.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        x = self.sigmoid(x)  # sigmoid to output
        return x

# initialize model, loss function, and optimizer
model = SimpleCNN().to(device)
criterion = nn.BCELoss()  # using binary cross-entropy loss since we have multi-label classification
optimizer = optim.Adam(model.parameters(), lr=0.001)

# calculate accuracy
def multi_label_accuracy(outputs, labels, threshold=0.5):
    outputs = (outputs > threshold).float()
    return torch.mean(((outputs == labels).float()).mean(dim=1)).item()
```

# Model Training and Validation

```python
# training loop
num_epochs = 5
batch_print_interval = 100  # interval for printing batch progress
print_interval = 1  # interval for printing epoch stats

for epoch in range(num_epochs):
    model.train()  # set model to training mode
    train_loss = 0.0
    train_accuracy = 0.0
    
    # Iterate over data
    for i, (images, labels) in enumerate(train_loader):
        images, labels = images.to(device), labels.to(device)
        
        optimizer.zero_grad()   # zero the parameter gradients
        
        # forward pass and loss calculation
        outputs = model(images)
        loss = criterion(outputs, labels)

        # backward pass and optimize
        loss.backward()
        optimizer.step()

        # accuracy
        batch_acc = multi_label_accuracy(outputs, labels)
        
        # update running training loss and accuracy
        train_loss += loss.item() * images.size(0)
        train_accuracy += batch_acc
        
        # print progress
        if (i + 1) % batch_print_interval == 0:
            print(f'Epoch [{epoch + 1}/{num_epochs}], Batch [{i + 1}/{len(train_loader)}], '
                  f'Loss: {loss.item():.4f}, Batch Accuracy: {batch_acc:.4f}')

    # avg loss and accuracy
    train_loss /= len(train_loader.dataset)
    train_accuracy /= len(train_loader.dataset)
    
    # val loop
    model.eval()  # 'eval' mode
    val_loss = 0.0
    val_accuracy = 0.0
    
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)

            # accuracy
            batch_acc = multi_label_accuracy(outputs, labels)
            
            # update running val loss and accuracy
            val_loss += loss.item() * images.size(0)
            val_accuracy += batch_acc
    
    # avg loss and accuracy
    val_loss /= len(val_loader.dataset)
    val_accuracy /= len(val_loader.dataset)
    
    # training and validation stats
    if (epoch + 1) % print_interval == 0:
        print(f'Epoch [{epoch + 1}/{num_epochs}], Training Loss: {train_loss:.4f}, '
              f'Training Accuracy: {train_accuracy:.4f}, Validation Loss: {val_loss:.4f}, '
              f'Validation Accuracy: {val_accuracy:.4f}')
```

```python
def dynamic_threshold_predictions(outputs, actual_labels):
    """
    This function adjusts the predicted labels based on the actual labels count for each image.
    For each image, it selects the top 'N' predictions, where 'N' is the number of actual labels.
    """
    all_predicted_labels = []

    for i in range(len(actual_labels)):
        actual_labels_count = int(actual_labels[i].sum().item())
        if actual_labels_count == 0:
            # if no labels -> predict none
            predicted = torch.zeros_like(outputs[i])
        else:
            # select the top N 'actual_labels_count' predictions
            _, top_indices = torch.topk(outputs[i], actual_labels_count)
            predicted = torch.zeros_like(outputs[i])
            predicted[top_indices] = 1

        all_predicted_labels.append(predicted)

    return torch.stack(all_predicted_labels)

def calculate_accuracy(predicted_labels, actual_labels):
    correct_predictions = 0
    total_predictions = 0
    for i in range(len(actual_labels)):
        actual_labels_count = int(actual_labels[i].sum().item())
        if actual_labels_count > 0:
            _, top_indices = torch.topk(actual_labels[i], actual_labels_count)
            correct_predictions += predicted_labels[i][top_indices].sum().item()
            total_predictions += actual_labels_count
    if total_predictions == 0:
        return 0  # special case -> no predictions (avoid division by 0)
    return correct_predictions / total_predictions

# eval loop w dynamic label prediction
model.eval()
val_loss = 0.0
val_corrects = 0
val_samples = 0

with torch.no_grad():
    for images, labels in val_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        loss = criterion(outputs, labels)
        
        predicted_labels = dynamic_threshold_predictions(outputs, labels)
        
        val_loss += loss.item() * images.size(0)
        
        batch_accuracy = calculate_accuracy(predicted_labels, labels)
        val_corrects += batch_accuracy * images.size(0)
        val_samples += images.size(0)

val_loss /= len(val_loader.dataset)
val_accuracy = val_corrects / val_samples

print(f'Validation Loss: {val_loss:.4f}, Validation Accuracy: {val_accuracy:.4f}')
```

```python
# update visualization loop to use dynamic threshold predictions
model.eval()
all_images = []
all_predicted_labels = []
all_actual_labels = []

with torch.no_grad():
    for batch_idx, (images, labels) in enumerate(val_loader):
        if len(all_images) * images.size(0) >= 20:
            break
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        predicted_labels = dynamic_threshold_predictions(outputs, labels)
        all_images.append(images)
        all_predicted_labels.append(predicted_labels)
        all_actual_labels.append(labels)

# convert all lists to tensors (for visualization)
all_images_tensor = torch.cat(all_images)
all_predicted_labels_tensor = torch.cat(all_predicted_labels)
all_actual_labels_tensor = torch.cat(all_actual_labels)

# convert to np arrays
all_images_np = all_images_tensor.cpu().numpy()
all_predicted_labels_np = all_predicted_labels_tensor.cpu().numpy()
all_actual_labels_np = all_actual_labels_tensor.cpu().numpy()

# plot the top 20 images w predicted AND actual labels
fig, axes = plt.subplots(20, 1, figsize=(5, 100))
for i in range(20):
    ax = axes[i]
    ax.imshow(all_images_np[i].transpose(1, 2, 0), cmap='gray')
    ax.axis('off')
    predicted_labels = [disease_categories[j] for j in range(len(disease_categories)) if all_predicted_labels_np[i][j] == 1]
    actual_labels = [disease_categories[j] for j in range(len(disease_categories)) if all_actual_labels_np[i][j] == 1]
    ax.set_title(f'Predicted: {predicted_labels}\nActual: {actual_labels}')

plt.tight_layout()
plt.show()
```
