```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from collections import Counter
from nltk.corpus import stopwords
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
def generate_summary(text,n):
    sentences = sent_tokenize(text)
    stop_words = set(stopwords.words('english'))
    words = [word.lower() for word in word_tokenize(text) if word.lower() not in stop_words and word.isalnum()]
    word_freq = Counter(words)
    sentence_scores = {}
    for sentence in sentences:
        sentence_words = [word.lower() for word in word_tokenize(sentence) if word.lower() not in stop_words and word.isalnum()]
        sentence_score = sum([word_freq[word] for word in sentence_words])
    if len(sentence_words) < 20:
        sentence_scores[sentence] = sentence_score\
        summary_sentences = sorted(sentence_scores, key=lambda x: sentence_scores[x], reverse=True)[:n]
        summary = ' '.join(summary_sentences)
        return summary
```

```python
import pandas as pd
from summarizer import Summarizer

def extractive_summarization(text):
    summarizer = Summarizer()
    summary = summarizer(text)
    return summary

def summarize_calls(file_path, output_csv_path):
    df = pd.read_excel(file_path)

    # Create a new column for summaries
    df['Summary'] = df['text'].apply(extractive_summarization)

    # Save the DataFrame with summaries to a new CSV file
    df.to_csv(output_csv_path, index=False)

# Example usage
file_path = 'training_dataset.xlsx'
output_csv_path = 'summarized_calls.csv'
summarize_calls(file_path, output_csv_path)
```

```python
import pandas as pd
from transformers import pipeline

def xlnet_summarization(text):
    summarizer = pipeline('summarization', model='facebook/bart-large-cnn')
    summary = summarizer(text, max_length=150, min_length=50, length_penalty=2.0)
    return summary[0]['summary_text']

def summarize_excel(input_excel, output_excel):
    # Read the Excel file into a pandas DataFrame
    df = pd.read_excel(input_excel)

    # Apply XLNet summarization to each row in the 'Transcript' column
    df['Summary'] = df['text'].apply(xlnet_summarization)

    # Save the DataFrame with summaries to a new Excel file
    df.to_excel(output_excel, index=False)

# Example usage
input_excel_path = 'training_dataset.xlsx'
output_excel_path = 'summarized_output.xlsx'
summarize_excel(input_excel_path, output_excel_path)
```

```python
import pandas as pd
from summa import keywords
from summa import summarizer

def textrank_extractive_summarization(text, num_sentences=3):
    # Extract keywords using TextRank
    extracted_keywords = keywords.keywords(text)

    # Use TextRank for extractive summarization
    summary = summarizer.summarize(text, words=num_sentences*50)  # Adjust words parameter based on your desired summary length

    return summary, extracted_keywords

def label_excel_with_textrank(input_excel_path, output_excel_path, num_sentences=3):
    # Read the Excel file
    df = pd.read_excel(input_excel_path)

    # Create new columns for summary and keywords
    df['Summary'] = ''
    df['Keywords'] = ''

    # Apply TextRank extractive summarization and labeling to each row
    for index, row in df.iterrows():
        text = str(row['text'])  # Assuming your Excel column is named 'Text'
        summary, keywords = textrank_extractive_summarization(text, num_sentences)
        df.at[index, 'Summary'] = summary
        df.at[index, 'Keywords'] = keywords

    # Save the labeled Excel file
    df.to_excel(output_excel_path, index=False)

# Example usage
input_excel_path = 'training_dataset.xlsx'  # Replace with your input Excel file path
output_excel_path = 'output_data_labeled.xlsx'  # Replace with your desired output Excel file path

label_excel_with_textrank(input_excel_path, output_excel_path)
```
